## **SPLIT THE DATA **

```python
X = df2.drop("student_marks", axis = "columns")
y = df2.drop("study_hours", axis = "columns")
```

```python
print("shape of X = ", X.shape)
print("shape of y = ", y.shape)
```

## **TRAIN AND TEST DATA**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size = 0.2, random_state=51)
```

```python
print("shape of X_train = ", X_train.shape)
print("shape of y_train = ", y_train.shape)
print("shape of X_test = ", X_test.shape)
print("shape of y_test = ", y_test.shape)
```

## **TRAINING MODEL**

```python
 from sklearn.linear_model import LinearRegression
 lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
#y=mx+c
lr.coef_  # m
```

```python
lr.intercept_#c
```

```python
m = 3.93
c = 50.44
y  = m * 4 + c  #if you study 4 hours thenyou will get 66.16 percentage using this equation
y
```

```python
lr.predict([[4]])[0][0].round(2)
```

```python
y_pred  = lr.predict(X_test) # predict unseeni.test data
y_pred
```

```python
#check actual y test with y pred :
pd.DataFrame(np.c_[X_test, y_test, y_pred], columns = ["study_hours", "student_marks_original","student_marks_predicted"])
```

## **EVALUATE MODEL PERFORMANCE:**

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_pred)

mse = mean_squared_error(y_test, y_pred)

rmse = np.sqrt(mse)

r2 = r2_score(y_test, y_pred)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```

```python
lr.score(X_test,y_test)   # r2 = r2_score(y_test, y_pred)  both same  and lr.score(X_test,y_test)one is internaly of lr model
```

```python
lr.score(X_train,y_train) # you can gain insights into how well your model generalizes to unseen data
#  want the R-squared score to be high on both the training and test datasets,
# indicating that your model is accurately capturing the underlying patterns without overfitting
```
